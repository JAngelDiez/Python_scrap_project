```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

tables = soup.find_all("div", class_="D(tbr)")

# Extracción de datos
revenue = []
for table in tables:
    rows = table.find_all("div", class_="D(tbc)")  # Modificar según la estructura
    for row in rows:
        revenue.append(row.get_text())
        
tesla_revenue_df = pd.DataFrame(revenue)

# Mostrar las últimas cinco filas
print(tesla_revenue_df.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ejemplo de DataFrame con datos de TSLA
# Reemplaza esta línea con el código para leer tus datos si ya los tienes en un archivo o DataFrame
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Close': [650, 652, 655, 660, 662]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Configurar el gráfico
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', color='b')
plt.title('Tesla (TSLA) Stock Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def make_graph(data, title):
    """
    Crea un gráfico de líneas de los datos proporcionados.
    
    :param data: DataFrame que debe contener al menos dos columnas: 'Date' y 'Close'.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], marker='o', linestyle='-', color='b')
    plt.title('GAMESTOP graphic')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Ejemplo de uso
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Close': [130, 135, 140, 145, 150]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Llamada a la función
make_graph(df, 'GameStop (GME) Stock Prices')
```
